```python
# Dependencies
import pandas as pd


from sqlalchemy import create_engine
```

```python
# Get the data!!!
data = pd.read_csv('./divorces_2000-2015_final_db.csv')
data
```

```python
data = data.dropna(subset=['type_of_divorce'])
data
```

```python
data
```

```python
# Create database connection
connection_string= 'postgresql+psycopg2://postgres:jesfer2011@localhost:5432/challenge3'
engine = create_engine(connection_string)
```

```python
data.to_sql('divorces', engine, index=False)
```

```python
pd.read_sql('select * from divorces', engine) 
```

```python
response = pd.read_sql("select zodiac_sign_man, count(type_of_divorce) from divorces where type_of_divorce = 'Voluntario' group by zodiac_sign_man", engine)
response = {
        'labels1': response['zodiac_sign_man'].to_list(),
        'values1': response['count'].to_list(),
    }
response
```
